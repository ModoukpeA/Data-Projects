# Visualisation Using Matplotlib Libray (A company sales dataset)

```python
import pandas  as pd
import matplotlib.pyplot as plt
company_sales = pd.read_csv("company_sales_data.csv")
company_sales
```

According to the displayed content of this dataset, as you can see, some conclusions can be drawn :
- From January to December, from month 1 to month 12 (a whole year), the dataset contains the sales of each product per month, the total units of each product sold and finally the total profit of all the sales per month.
- All the values are numeric (which makes the calculations easily performed)
Now, let's make some data preparation and #vizzes based on each question.

# Read Total profit of all months and show it using a line plot

```python
total_profit_per_month = company_sales[["month_number","total_profit"]]
total_profit_per_month
```

Above, we display the total profit made on all the products sold per month. As you can notice, the month in which the company makes the highest profit is the month of November. This can be due to the fact that it's almost the end of year.

```python
x = company_sales["month_number"]
y = company_sales["total_profit"]
plt.plot(x, y, 'b')
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(x)
plt.legend(title = "Total Profit per month")
plt.title("Total Profit per month")
plt.show()
```

On the displayed plot, the highest point is clearly shown. And you can see that it really happens in the 11th month of the year.

# Read all product sales data and show it  using a multi line plot

```python
month_number = company_sales["month_number"]
facecream = company_sales["facecream"]
facewash = company_sales["facewash"]
toothpaste = company_sales["toothpaste"]
bathingsoap = company_sales["bathingsoap"]
shampoo = company_sales["shampoo"]
moisturizer = company_sales["moisturizer"]

plt.plot(month_number,facecream,label="Face Cream Sales Data")
plt.plot(month_number,facewash,label="Face Wash Sales Data")
plt.plot(month_number,toothpaste,label="ToothPaste Sales Data")
plt.plot(month_number,bathingsoap,label="Bathing Soap Sales Data")
plt.plot(month_number,shampoo,label="Shampoo Sales Data")
plt.plot(month_number,moisturizer,label="Moisturizer Sales Data")

plt.xlabel("Month Number")
plt.ylabel("Sales Units in number")
plt.xticks(month_number)
plt.legend()
plt.title("Sales data")
plt.show()
```

# Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(month_number,toothpaste,label="ToothPaste Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales Units in number")
plt.xticks(month_number)
plt.legend()
plt.title("Toothpaste sales data")
plt.grid(linestyle = '--')
plt.show()
```

# Read face cream and facewash product sales data and show it using the bar chart

```python
plt.bar(month_number,facecream,label="Face Cream Sales Data")
plt.bar(month_number,facewash,label="Face Wash Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales Units in number")
plt.xticks(month_number)
plt.legend()
plt.title("FaceCream and FaceWash sales data")
plt.grid(linestyle = '--')
plt.show()
```

# Calculate total sale data for last year for each product and show it using a Pie chart

```python
company_sales
```

```python
products_sales = company_sales[["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]]
products_sales
```

```python
import numpy as np
each_product_total_sales = np.array(products_sales.sum(axis=0))
each_product_total_sales
```

```python
x=np.array(["FaceCream","FaceWash","ToothPaste","BathingSoap","Shampoo","Moisturizer"])
plt.pie(each_product_total_sales, labels=x, autopct='%1.1f%%')
plt.legend()
plt.title("Sales Data")
plt.show()
```

Bathing Soaps are really purchased by clients during the year in this company. Next, we have the toothpaste, the FaceCream, the Shampoo. Finally, it seems that people barely buy FaceWash and Moisturizer in this company. This can be caused by the quality of these products (FaceCream and Moisturizer) in the company. They are either too expensive or have a low quality.
